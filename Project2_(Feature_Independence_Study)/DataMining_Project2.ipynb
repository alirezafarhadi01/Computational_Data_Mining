```python
# ================================
# Step 2 – Initial Collinearity Analysis
# ================================
# Tasks:
#  1. Compute covariance / correlation structure of features.
#  2. Visualize with heatmaps.
#  3. Print a basic analysis of feature dependencies.
#
# This cell:
#  - Uses correlation matrices (scale-invariant and standard for collinearity).
#  - Creates a heatmap per dataset.
#  - Prints the top most correlated feature pairs (by |corr|) as documentation.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ---------------------------------------------------------
# Ensure datasets from Step 1 are available; otherwise, reload them
# ---------------------------------------------------------
try:
    X_cls
    X_reg
    X_clust
except NameError:
    from sklearn.datasets import load_breast_cancer, fetch_california_housing, load_iris
    X_cls = load_breast_cancer(as_frame=True).data
    X_reg = fetch_california_housing(as_frame=True).data
    X_clust = load_iris(as_frame=True).data

sns.set(style="white", font_scale=0.9)

def analyze_collinearity(X: pd.DataFrame, name: str, top_k: int = 10):
    """
    Compute correlation matrix, show heatmap, and print top correlated feature pairs.
    """
    print(f"\n{'='*80}")
    print(f"Dataset: {name}")
    print(f"Shape (samples, features): {X.shape}")

    # ---------------------------------------------
    # 1) Correlation matrix (for collinearity)
    # ---------------------------------------------
    corr = X.corr()  # Pearson correlation by default

    # ---------------------------------------------
    # 2) Heatmap visualization
    # ---------------------------------------------
    # Figure size scales mildly with number of features
    plt.figure(figsize=(min(1.2 * X.shape[1], 12), min(1.0 * X.shape[1], 10)))
    sns.heatmap(
        corr,
        annot=False,
        cmap="coolwarm",
        center=0,
        square=True,
        cbar_kws={"shrink": 0.7}
    )
    plt.title(f"Feature Correlation Heatmap – {name}")
    plt.tight_layout()
    plt.show()

    # ---------------------------------------------
    # 3) Numerical summary of dependency (top |corr|)
    # ---------------------------------------------
    corr_abs = corr.abs()

    # Take only upper-triangular (without diagonal) to avoid duplicates/self-corr
    mask = np.triu(np.ones_like(corr_abs, dtype=bool), k=1)
    corr_pairs = corr_abs.where(mask).stack().sort_values(ascending=False)

    if corr_pairs.empty:
        print("Not enough features to compute pairwise correlations.")
        return

    max_corr = corr_pairs.iloc[0]
    print(f"Maximum absolute correlation between two distinct features: {max_corr:.3f}")

    print(f"\nTop {min(top_k, len(corr_pairs))} most correlated feature pairs (by |corr|):")
    for (f1, f2), val in corr_pairs.iloc[:top_k].items():
        if val >= 0.9:
            relation = "very high"
        elif val >= 0.7:
            relation = "high"
        elif val >= 0.4:
            relation = "moderate"
        else:
            relation = "low"
        print(f"  {f1} ↔ {f2}: |corr| = {val:.3f}  ({relation} dependency)")

    # ---------------------------------------------
    # 4) Short textual guideline for your report
    # ---------------------------------------------
    print("\nInterpretation note:")
    print(
        "- Very high / high correlations (|corr| ≥ 0.7) indicate strong collinearity: \n"
        "  those features carry very similar information and may cause instability in\n"
        "  linear models and slow or unstable optimization.\n"
        "- Moderate correlations (0.4 ≤ |corr| < 0.7) suggest partial redundancy.\n"
        "- Low correlations (|corr| < 0.4) indicate more independent information\n"
        "  across features, which is usually better for stable coefficient estimation\n"
        "  and well-conditioned optimization problems."
    )

# Run collinearity analysis for each dataset
analyze_collinearity(X_cls,   "Classification – Breast Cancer")
analyze_collinearity(X_reg,   "Regression – California Housing")
analyze_collinearity(X_clust, "Clustering – Iris")

print("\nStep 2 completed: correlation matrices computed, heatmaps visualized, and basic collinearity analysis printed.")
```

```python
# ================================
# Step 4 – Applying Basic Feature Selection Methods
# ================================
# Tasks:
#  1. Apply SelectKBest (e.g., f_regression / f_classif) to select top features.
#  2. Apply RFE (Recursive Feature Elimination) with suitable estimators.
#  3. Work on the ORIGINAL feature space (not PCA/ICA/SVD features).
#
# This cell:
#  - Ensures X_cls, y_cls, X_reg, y_reg, X_clust, y_clust exist.
#  - For each dataset:
#       * Chooses k = half of the original features (at least 1).
#       * Runs SelectKBest.
#       * Runs RFE with an appropriate estimator.
#       * Stores selected feature subsets in variables for later steps.
#  - Prints which features were selected.

import numpy as np
import pandas as pd

from sklearn.datasets import load_breast_cancer, fetch_california_housing, load_iris
from sklearn.feature_selection import SelectKBest, f_regression, f_classif, RFE
from sklearn.linear_model import LinearRegression, LogisticRegression

# ---------------------------------------------------------
# Ensure original datasets are available (from Step 1); otherwise reload
# ---------------------------------------------------------
try:
    X_cls, y_cls
    X_reg, y_reg
    X_clust, y_clust
except NameError:
    bc = load_breast_cancer(as_frame=True)
    cal = fetch_california_housing(as_frame=True)
    iris = load_iris(as_frame=True)

    X_cls = bc.data
    y_cls = bc.target

    X_reg = cal.data
    y_reg = cal.target

    X_clust = iris.data
    y_clust = iris.target   # used here as labels to measure feature relevance

# ---------------------------------------------------------
# Helper function for feature selection
# ---------------------------------------------------------
def apply_feature_selection(name, X: pd.DataFrame, y: np.ndarray, problem_type: str):
    """
    Apply SelectKBest + RFE to select a subset of features.

    problem_type: 'classification' or 'regression'
    """
    print("=" * 80)
    print(f"Dataset: {name}")
    print(f"Original shape: {X.shape}")

    n_features = X.shape[1]
    # Choose k as half of the features (rounded down), but at least 1
    k = max(1, n_features // 2)
    print(f"Selecting k = {k} features (out of {n_features})")

    # -------------------------
    # 1) SelectKBest
    # -------------------------
    if problem_type == "regression":
        score_func = f_regression
    else:  # classification
        score_func = f_classif

    skb = SelectKBest(score_func=score_func, k=k)
    X_kbest = skb.fit_transform(X, y)
    mask_kbest = skb.get_support()
    selected_features_kbest = X.columns[mask_kbest]

    print("\nSelectKBest – selected features:")
    for f in selected_features_kbest:
        print(f"  - {f}")
    print(f"Resulting shape (SelectKBest): {X_kbest.shape}")

    # -------------------------
    # 2) RFE (Recursive Feature Elimination)
    # -------------------------
    if problem_type == "regression":
        estimator = LinearRegression()
    else:
        # LogisticRegression as a simple linear classifier with coefficients
        estimator = LogisticRegression(max_iter=2000)

    rfe = RFE(estimator=estimator, n_features_to_select=k, step=1)
    X_rfe = rfe.fit_transform(X, y)
    mask_rfe = rfe.support_
    selected_features_rfe = X.columns[mask_rfe]

    print("\nRFE – selected features:")
    for f in selected_features_rfe:
        print(f"  - {f}")
    print(f"Resulting shape (RFE): {X_rfe.shape}")

    # -------------------------
    # 3) Short note for the report
    # -------------------------
    print("\nSummary note:")
    print(
        "- SelectKBest chooses features based purely on univariate statistical tests\n"
        "  between each individual feature and the target.\n"
        "- RFE uses the estimator's coefficients to iteratively remove less important\n"
        "  features, capturing multivariate interactions better than univariate tests.\n"
    )

    return X_kbest, X_rfe, list(selected_features_kbest), list(selected_features_rfe)


# ---------------------------------------------------------
# Apply to each dataset
# ---------------------------------------------------------

# 1) Classification – Breast Cancer
X_cls_kbest, X_cls_rfe, cls_kbest_features, cls_rfe_features = apply_feature_selection(
    name="Classification – Breast Cancer",
    X=X_cls,
    y=y_cls,
    problem_type="classification",
)

# 2) Regression – California Housing
X_reg_kbest, X_reg_rfe, reg_kbest_features, reg_rfe_features = apply_feature_selection(
    name="Regression – California Housing",
    X=X_reg,
    y=y_reg,
    problem_type="regression",
)

# 3) Clustering – Iris (using species labels as pseudo-targets for feature relevance)
X_clust_kbest, X_clust_rfe, clust_kbest_features, clust_rfe_features = apply_feature_selection(
    name="Clustering – Iris (using class labels for feature relevance)",
    X=X_clust,
    y=y_clust,
    problem_type="classification",
)

print("\nStep 4 completed: SelectKBest and RFE applied; selected feature subsets stored in:")
print("  - X_cls_kbest, X_cls_rfe")
print("  - X_reg_kbest, X_reg_rfe")
print("  - X_clust_kbest, X_clust_rfe")
print("Feature name lists stored in:")
print("  - cls_kbest_features, cls_rfe_features")
print("  - reg_kbest_features, reg_rfe_features")
print("  - clust_kbest_features, clust_rfe_features")
```

```python
# ============================================
# Step 5 – Model Training and Optimization Analysis
# ============================================
# Assumes previous steps defined:
#   - X_reg, y_reg, X_reg_scaled, X_reg_pca, X_reg_kbest
#   - X_clust, X_clust_scaled, X_clust_pca
#   - X_cls, y_cls, X_cls_scaled, X_cls_pca, X_cls_kbest
#
# From:
#   - Step 1: original datasets
#   - Step 3: PCA + scaled data
#   - Step 4: feature-selected data (SelectKBest)
#
# This step:
#   (a) Regression – LinearRegression + SGDRegressor
#   (b) Clustering – KMeans (original vs PCA)
#   (c) Classification – KNN + RandomForest
# ============================================

import numpy as np
import time
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression, SGDRegressor
from sklearn.metrics import mean_squared_error, accuracy_score
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler

# ---------------------------------------------------------
# Safety checks – if something is missing, raise a clear error
# ---------------------------------------------------------
required_vars = [
    "X_reg", "y_reg", "X_reg_scaled", "X_reg_pca", "X_reg_kbest",
    "X_clust", "X_clust_scaled", "X_clust_pca",
    "X_cls", "y_cls", "X_cls_scaled", "X_cls_pca", "X_cls_kbest"
]

missing = [v for v in required_vars if v not in globals()]
if missing:
    raise RuntimeError(
        "The following variables are missing. "
        "Make sure you have run Steps 1–4 before Step 5:\n" + ", ".join(missing)
    )

# ---------------------------------------------------------
# Prepare scaled feature-selected versions where needed
# ---------------------------------------------------------

# Regression – scale selected features (SelectKBest)
scaler_reg_kbest = StandardScaler()
X_reg_kbest_scaled = scaler_reg_kbest.fit_transform(X_reg_kbest)

# Classification – scale selected features (SelectKBest)
scaler_cls_kbest = StandardScaler()
X_cls_kbest_scaled = scaler_cls_kbest.fit_transform(X_cls_kbest)

# For clustering we only use original (scaled) and PCA, not feature selection.


# =========================================================
# (a) Regression Problem – Analytical vs Derivative-Based
#       Dataset: California Housing
# =========================================================

print("\n" + "#" * 80)
print("(a) Regression – California Housing (LinearRegression + SGDRegressor)")
print("#" * 80 + "\n")

# -----------------------------
# Train/Test split (consistent across representations)
# -----------------------------
(
    X_reg_train_orig, X_reg_test_orig,
    X_reg_train_pca,  X_reg_test_pca,
    X_reg_train_kb,   X_reg_test_kb,
    y_reg_train,      y_reg_test
) = train_test_split(
    X_reg_scaled, X_reg_pca, X_reg_kbest_scaled, y_reg,
    test_size=0.2,
    random_state=42
)

# -----------------------------
# (a.1) Model 1 – LinearRegression (Analytical / Closed-form)
# -----------------------------
lin_orig = LinearRegression()
lin_orig.fit(X_reg_train_orig, y_reg_train)

lin_pca = LinearRegression()
lin_pca.fit(X_reg_train_pca, y_reg_train)

lin_kb = LinearRegression()
lin_kb.fit(X_reg_train_kb, y_reg_train)

# Coefficient statistics
coef_orig = lin_orig.coef_
coef_pca = lin_pca.coef_
coef_kb = lin_kb.coef_

def summarize_coefs(name, coefs):
    print(f"  [{name}]")
    print(f"    Number of coefficients: {coefs.shape[0]}")
    print(f"    L2 norm of coefficients: {np.linalg.norm(coefs):.4f}")
    print(f"    Mean |coef|: {np.mean(np.abs(coefs)):.4f}")
    print(f"    Max |coef|: {np.max(np.abs(coefs)):.4f}")

print("LinearRegression – Coefficient stability comparison:")
summarize_coefs("Original (scaled)", coef_orig)
summarize_coefs("PCA features", coef_pca)
summarize_coefs("Selected features (SelectKBest, scaled)", coef_kb)

# Errors
def mse_report(model, X_tr, y_tr, X_te, y_te, label):
    y_tr_pred = model.predict(X_tr)
    y_te_pred = model.predict(X_te)
    tr_mse = mean_squared_error(y_tr, y_tr_pred)
    te_mse = mean_squared_error(y_te, y_te_pred)
    print(f"  {label}: Train MSE = {tr_mse:.4f}, Test MSE = {te_mse:.4f}")
    return tr_mse, te_mse

print("\nLinearRegression – Train/Test MSE:")
mse_report(lin_orig, X_reg_train_orig, y_reg_train, X_reg_test_orig, y_reg_test, "Original (scaled)")
mse_report(lin_pca,  X_reg_train_pca,  y_reg_train, X_reg_test_pca,  y_reg_test, "PCA features")
mse_report(lin_kb,   X_reg_train_kb,   y_reg_train, X_reg_test_kb,   y_reg_test, "Selected (SelectKBest)")

print("\nInterpretation hint for your report:")
print(
    "- Large magnitude and highly variable coefficients on the original (collinear) features\n"
    "  indicate sensitivity of the solution to small changes in the data, a sign of multicollinearity.\n"
    "- PCA features are orthogonal, so coefficients tend to be more balanced and stable,\n"
    "  since each component captures independent variance."
)

# -----------------------------
# (a.2) Model 2 – SGDRegressor (Derivative-Based)
# -----------------------------
sgd_orig = SGDRegressor(
    max_iter=2000,
    tol=1e-4,
    random_state=42,
    learning_rate="invscaling",
    eta0=0.01
)
sgd_orig.fit(X_reg_train_orig, y_reg_train)

sgd_pca = SGDRegressor(
    max_iter=2000,
    tol=1e-4,
    random_state=42,
    learning_rate="invscaling",
    eta0=0.01
)
sgd_pca.fit(X_reg_train_pca, y_reg_train)

print("\nSGDRegressor – Basic convergence stats (scikit-learn fit):")
print(f"  Original (scaled): n_iter_ = {sgd_orig.n_iter_}, Train MSE = "
      f"{mean_squared_error(y_reg_train, sgd_orig.predict(X_reg_train_orig)):.4f}")
print(f"  PCA features:      n_iter_ = {sgd_pca.n_iter_}, Train MSE = "
      f"{mean_squared_error(y_reg_train, sgd_pca.predict(X_reg_train_pca)):.4f}")

# --- Manual training loop for convergence curves (for plotting) ---
def sgd_training_curve(X_train, y_train, X_val, y_val, max_epochs=100, lr=0.01, alpha=0.0001, random_state=42):
    model = SGDRegressor(
        max_iter=1,        # one epoch at a time
        tol=None,
        learning_rate="constant",
        eta0=lr,
        alpha=alpha,
        random_state=random_state,
        warm_start=True
    )

    train_losses = []
    val_losses = []

    for epoch in range(max_epochs):
        model.partial_fit(X_train, y_train)
        y_tr_pred = model.predict(X_train)
        y_val_pred = model.predict(X_val)

        train_losses.append(mean_squared_error(y_train, y_tr_pred))
        val_losses.append(mean_squared_error(y_val, y_val_pred))

    return model, np.array(train_losses), np.array(val_losses)

# Use 20% of the training set as a "validation" set for curves
(
    X_tr_orig_sub, X_val_orig,
    X_tr_pca_sub,  X_val_pca,
    y_tr_sub,      y_val_sub
) = train_test_split(
    X_reg_train_orig, X_reg_train_pca, y_reg_train,
    test_size=0.2,
    random_state=42
)

max_epochs = 100

model_sgd_orig_curve, tr_loss_orig, val_loss_orig = sgd_training_curve(
    X_tr_orig_sub, y_tr_sub, X_val_orig, y_val_sub,
    max_epochs=max_epochs, lr=0.01, alpha=0.0001, random_state=42
)

model_sgd_pca_curve, tr_loss_pca, val_loss_pca = sgd_training_curve(
    X_tr_pca_sub, y_tr_sub, X_val_pca, y_val_sub,
    max_epochs=max_epochs, lr=0.01, alpha=0.0001, random_state=42
)

# Plot convergence curves (Required Output)
epochs = np.arange(1, max_epochs + 1)

plt.figure(figsize=(7, 4))
plt.plot(epochs, val_loss_orig, label="SGD on Original (scaled)")
plt.plot(epochs, val_loss_pca, label="SGD on PCA features")
plt.xlabel("Epoch")
plt.ylabel("Validation MSE")
plt.title("SGDRegressor Convergence – Original vs PCA Features")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("\nObservation hint for your report:")
print(
    "- You should typically observe that the PCA-based model reaches a low validation loss\n"
    "  in fewer epochs (smoother, faster convergence), thanks to the orthogonalized feature\n"
    "  space, while the original collinear space may converge more slowly or noisily."
)


# =========================================================
# (b) Clustering Problem – KMeans (Iris)
# =========================================================

print("\n" + "#" * 80)
print("(b) Clustering – Iris (KMeans on Original vs PCA)")
print("#" * 80 + "\n")

n_clusters = 3  # Iris has 3 species

# Original (scaled) data
X_clust_orig = X_clust_scaled

# PCA-reduced data (2 or 3 components)
n_pca_clust = min(3, X_clust_pca.shape[1])
X_clust_pca_reduced = X_clust_pca[:, :n_pca_clust]

# --- KMeans on original data ---
kmeans_orig = KMeans(n_clusters=n_clusters, n_init=10, random_state=42)
start = time.perf_counter()
kmeans_orig.fit(X_clust_orig)
time_orig = time.perf_counter() - start

inertia_orig = kmeans_orig.inertia_
silhouette_orig = silhouette_score(X_clust_orig, kmeans_orig.labels_)

# --- KMeans on PCA-reduced data ---
kmeans_pca = KMeans(n_clusters=n_clusters, n_init=10, random_state=42)
start = time.perf_counter()
kmeans_pca.fit(X_clust_pca_reduced)
time_pca = time.perf_counter() - start

inertia_pca = kmeans_pca.inertia_
silhouette_pca = silhouette_score(X_clust_pca_reduced, kmeans_pca.labels_)

print("KMeans results:")
print(f"  Original (scaled) data:")
print(f"    Inertia:         {inertia_orig:.4f}")
print(f"    Silhouette:      {silhouette_orig:.4f}")
print(f"    Iterations:      {kmeans_orig.n_iter_}")
print(f"    Computation time:{time_orig:.6f} seconds")

print(f"\n  PCA-reduced data (first {n_pca_clust} components):")
print(f"    Inertia:         {inertia_pca:.4f}")
print(f"    Silhouette:      {silhouette_pca:.4f}")
print(f"    Iterations:      {kmeans_pca.n_iter_}")
print(f"    Computation time:{time_pca:.6f} seconds")

print("\nInterpretation hint for your report:")
print(
    "- KMeans is distance-based and sensitive to the geometry of the feature space.\n"
    "- PCA often reduces noise and collinearity, so clustering in the PCA space can\n"
    "  lead to similar or better Silhouette scores with lower inertia and lower\n"
    "  computation time, especially in high-dimensional problems."
)


# =========================================================
# (c) Classification Problem – KNN and RandomForest (Breast Cancer)
# =========================================================

print("\n" + "#" * 80)
print("(c) Classification – Breast Cancer (KNN + RandomForest)")
print("#" * 80 + "\n")

# -----------------------------
# Train/Test split (consistent across representations)
# -----------------------------
(
    X_cls_train_orig, X_cls_test_orig,
    X_cls_train_pca,  X_cls_test_pca,
    X_cls_train_kb,   X_cls_test_kb,
    y_cls_train,      y_cls_test
) = train_test_split(
    X_cls_scaled, X_cls_pca, X_cls_kbest_scaled, y_cls,
    test_size=0.2,
    stratify=y_cls,
    random_state=42
)

# -----------------------------
# (c.1) KNN – distance-based model
# -----------------------------
knn = KNeighborsClassifier(n_neighbors=5)

# Original (scaled)
knn_orig = KNeighborsClassifier(n_neighbors=5)
knn_orig.fit(X_cls_train_orig, y_cls_train)
start = time.perf_counter()
y_pred_orig = knn_orig.predict(X_cls_test_orig)
pred_time_orig = time.perf_counter() - start
acc_orig = accuracy_score(y_cls_test, y_pred_orig)

# PCA
knn_pca = KNeighborsClassifier(n_neighbors=5)
knn_pca.fit(X_cls_train_pca, y_cls_train)
start = time.perf_counter()
y_pred_pca = knn_pca.predict(X_cls_test_pca)
pred_time_pca = time.perf_counter() - start
acc_pca = accuracy_score(y_cls_test, y_pred_pca)

# (Optional) Selected features – also distance-based, included for completeness
knn_kb = KNeighborsClassifier(n_neighbors=5)
knn_kb.fit(X_cls_train_kb, y_cls_train)
start = time.perf_counter()
y_pred_kb = knn_kb.predict(X_cls_test_kb)
pred_time_kb = time.perf_counter() - start
acc_kb = accuracy_score(y_cls_test, y_pred_kb)

print("KNN – Accuracy and prediction time:")
print(f"  Original (scaled):  Accuracy = {acc_orig:.4f}, Prediction time = {pred_time_orig:.6f} s")
print(f"  PCA features:       Accuracy = {acc_pca:.4f}, Prediction time = {pred_time_pca:.6f} s")
print(f"  Selected (KBest):   Accuracy = {acc_kb:.4f}, Prediction time = {pred_time_kb:.6f} s")

print("\nInterpretation hint for your report (KNN):")
print(
    "- KNN relies heavily on Euclidean distances. Reducing dimensionality with PCA can\n"
    "  both speed up prediction (fewer dimensions to compare) and sometimes improve\n"
    "  generalization if noisy or redundant features are removed.\n"
    "- Feature selection (KBest) keeps only a subset of original features and may balance\n"
    "  between performance and computational cost."
)

# -----------------------------
# (c.2) RandomForest – tree-based ensemble
# -----------------------------
rf_orig = RandomForestClassifier(
    n_estimators=200,
    random_state=42,
    n_jobs=-1
)
rf_orig.fit(X_cls_train_orig, y_cls_train)
acc_rf_orig = accuracy_score(y_cls_test, rf_orig.predict(X_cls_test_orig))

rf_pca = RandomForestClassifier(
    n_estimators=200,
    random_state=42,
    n_jobs=-1
)
rf_pca.fit(X_cls_train_pca, y_cls_train)
acc_rf_pca = accuracy_score(y_cls_test, rf_pca.predict(X_cls_test_pca))

rf_kb = RandomForestClassifier(
    n_estimators=200,
    random_state=42,
    n_jobs=-1
)
rf_kb.fit(X_cls_train_kb, y_cls_train)
acc_rf_kb = accuracy_score(y_cls_test, rf_kb.predict(X_cls_test_kb))

print("\nRandomForest – Accuracy across feature spaces:")
print(f"  Original (scaled):  Accuracy = {acc_rf_orig:.4f}")
print(f"  PCA features:       Accuracy = {acc_rf_pca:.4f}")
print(f"  Selected (KBest):   Accuracy = {acc_rf_kb:.4f}")

print("\nInterpretation hint for your report (RandomForest):")
print(
    "- Random Forests (and tree-based models) are generally more robust to multicollinearity\n"
    "  because each tree splits on single features and ensembles average over many trees.\n"
    "- As a result, the performance gap between original, PCA, and feature-selected spaces\n"
    "  is often smaller than for linear or distance-based models.\n"
    "- You can still observe whether PCA or feature selection simplifies the model without\n"
    "  significantly hurting accuracy."
)

print("\nStep 5 completed: all requested models trained and analyzed on original, PCA, and selected feature spaces.")
```

```python
# ============================================
# Step 6 – Analysis of Results and Visualizations
# ============================================
# This cell:
#   - Recomputes key metrics from the fitted models in Step 5
#   - Plots and compares:
#       * Regression errors and coefficient norms
#       * SGD convergence summary (final val loss)
#       * KMeans clustering metrics (inertia, silhouette, time)
#       * KNN accuracy and prediction time
#       * RandomForest accuracy across feature spaces
# ============================================

import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, accuracy_score

# ---------------------------------------------------------
# Safety check – require key variables from Step 5
# ---------------------------------------------------------
needed = [
    # Regression models and data
    "lin_orig", "lin_pca", "lin_kb",
    "sgd_orig", "sgd_pca",
    "X_reg_test_orig", "X_reg_test_pca", "X_reg_test_kb", "y_reg_test",
    # SGD convergence curves (from Step 5)
    "val_loss_orig", "val_loss_pca",
    # Clustering metrics
    "inertia_orig", "inertia_pca",
    "silhouette_orig", "silhouette_pca",
    "time_orig", "time_pca",
    # Classification models and data
    "knn_orig", "knn_pca", "knn_kb",
    "rf_orig", "rf_pca", "rf_kb",
    "X_cls_test_orig", "X_cls_test_pca", "X_cls_test_kb", "y_cls_test",
]

missing = [v for v in needed if v not in globals()]
if missing:
    raise RuntimeError(
        "These variables are missing. Make sure you ran Step 5 before Step 6:\n"
        + ", ".join(missing)
    )

# =========================================================
# 1) Regression – LinearRegression & SGDRegressor
# =========================================================

# --- LinearRegression: coefficient norms and test MSE ---
coef_orig = lin_orig.coef_
coef_pca  = lin_pca.coef_
coef_kb   = lin_kb.coef_

coef_norms = [
    np.linalg.norm(coef_orig),
    np.linalg.norm(coef_pca),
    np.linalg.norm(coef_kb)
]

lin_mse_test_orig = mean_squared_error(y_reg_test, lin_orig.predict(X_reg_test_orig))
lin_mse_test_pca  = mean_squared_error(y_reg_test, lin_pca.predict(X_reg_test_pca))
lin_mse_test_kb   = mean_squared_error(y_reg_test, lin_kb.predict(X_reg_test_kb))

lin_mses = [lin_mse_test_orig, lin_mse_test_pca, lin_mse_test_kb]

spaces_3 = ["Original", "PCA", "KBest"]

# --- Plot 1: Coefficient L2 norms for LinearRegression ---
plt.figure(figsize=(5, 3))
plt.bar(spaces_3, coef_norms)
plt.ylabel("L2 norm of coefficients")
plt.title("LinearRegression – Coefficient magnitude by feature space")
plt.tight_layout()
plt.show()

# --- Plot 2: Test MSE for LinearRegression ---
plt.figure(figsize=(5, 3))
plt.bar(spaces_3, lin_mses)
plt.ylabel("Test MSE")
plt.title("LinearRegression – Test error by feature space")
plt.tight_layout()
plt.show()

# --- SGDRegressor: test MSE and final validation loss ---
sgd_mse_test_orig = mean_squared_error(y_reg_test, sgd_orig.predict(X_reg_test_orig))
sgd_mse_test_pca  = mean_squared_error(y_reg_test, sgd_pca.predict(X_reg_test_pca))

sgd_test_mses = [sgd_mse_test_orig, sgd_mse_test_pca]
spaces_2 = ["Original", "PCA"]

# Final validation losses from convergence curves (Step 5)
sgd_val_final_orig = float(val_loss_orig[-1])
sgd_val_final_pca  = float(val_loss_pca[-1])
sgd_val_final = [sgd_val_final_orig, sgd_val_final_pca]

# --- Plot 3: SGDRegressor – Test MSE (Original vs PCA) ---
plt.figure(figsize=(5, 3))
plt.bar(spaces_2, sgd_test_mses)
plt.ylabel("Test MSE")
plt.title("SGDRegressor – Test error by feature space")
plt.tight_layout()
plt.show()

# --- Plot 4: SGDRegressor – Final validation loss (from curves) ---
plt.figure(figsize=(5, 3))
plt.bar(spaces_2, sgd_val_final)
plt.ylabel("Final validation MSE")
plt.title("SGDRegressor – Final validation loss (Original vs PCA)")
plt.tight_layout()
plt.show()

print("Regression summary (printed values):")
print(f"  LinearRegression Test MSE – Original: {lin_mse_test_orig:.4f}, "
      f"PCA: {lin_mse_test_pca:.4f}, KBest: {lin_mse_test_kb:.4f}")
print(f"  SGDRegressor Test MSE – Original: {sgd_mse_test_orig:.4f}, "
      f"PCA: {sgd_mse_test_pca:.4f}")
print(f"  SGDRegressor Final val MSE – Original: {sgd_val_final_orig:.4f}, "
      f"PCA: {sgd_val_final_pca:.4f}")
print()


# =========================================================
# 2) Clustering – KMeans metrics (Iris)
# =========================================================

inertia_vals    = [inertia_orig, inertia_pca]
silhouette_vals = [silhouette_orig, silhouette_pca]
time_vals       = [time_orig, time_pca]

fig, axes = plt.subplots(1, 3, figsize=(12, 3))

axes[0].bar(spaces_2, inertia_vals)
axes[0].set_title("KMeans – Inertia")
axes[0].set_ylabel("Inertia")

axes[1].bar(spaces_2, silhouette_vals)
axes[1].set_title("KMeans – Silhouette score")
axes[1].set_ylabel("Silhouette")

axes[2].bar(spaces_2, time_vals)
axes[2].set_title("KMeans – Computation time")
axes[2].set_ylabel("Time (seconds)")

for ax in axes:
    ax.set_xlabel("Feature space")

plt.suptitle("KMeans – Original vs PCA (Iris)")
plt.tight_layout()
plt.show()

print("KMeans summary (printed values):")
print(f"  Original – Inertia: {inertia_orig:.4f}, Silhouette: {silhouette_orig:.4f}, "
      f"Time: {time_orig:.6f} s")
print(f"  PCA      – Inertia: {inertia_pca:.4f}, Silhouette: {silhouette_pca:.4f}, "
      f"Time: {time_pca:.6f} s")
print()


# =========================================================
# 3) Classification – KNN and RandomForest (Breast Cancer)
# =========================================================

# --- Recompute accuracies and prediction times for KNN ---
import time as _time

# Original
start = _time.perf_counter()
y_pred_knn_orig = knn_orig.predict(X_cls_test_orig)
pred_time_knn_orig = _time.perf_counter() - start
acc_knn_orig = accuracy_score(y_cls_test, y_pred_knn_orig)

# PCA
start = _time.perf_counter()
y_pred_knn_pca = knn_pca.predict(X_cls_test_pca)
pred_time_knn_pca = _time.perf_counter() - start
acc_knn_pca = accuracy_score(y_cls_test, y_pred_knn_pca)

# KBest
start = _time.perf_counter()
y_pred_knn_kb = knn_kb.predict(X_cls_test_kb)
pred_time_knn_kb = _time.perf_counter() - start
acc_knn_kb = accuracy_score(y_cls_test, y_pred_knn_kb)

knn_accs = [acc_knn_orig, acc_knn_pca, acc_knn_kb]
knn_times = [pred_time_knn_orig, pred_time_knn_pca, pred_time_knn_kb]

# --- RandomForest accuracies ---
acc_rf_orig = accuracy_score(y_cls_test, rf_orig.predict(X_cls_test_orig))
acc_rf_pca  = accuracy_score(y_cls_test, rf_pca.predict(X_cls_test_pca))
acc_rf_kb   = accuracy_score(y_cls_test, rf_kb.predict(X_cls_test_kb))

rf_accs = [acc_rf_orig, acc_rf_pca, acc_rf_kb]

# --- Plot 5: KNN Accuracy ---
plt.figure(figsize=(5, 3))
plt.bar(spaces_3, knn_accs)
plt.ylabel("Accuracy")
plt.ylim(0, 1.0)
plt.title("KNN – Accuracy by feature space")
plt.tight_layout()
plt.show()

# --- Plot 6: KNN Prediction Time ---
plt.figure(figsize=(5, 3))
plt.bar(spaces_3, knn_times)
plt.ylabel("Prediction time (seconds)")
plt.title("KNN – Prediction time by feature space")
plt.tight_layout()
plt.show()

print("KNN summary (printed values):")
print(f"  Original – Accuracy: {acc_knn_orig:.4f}, Prediction time: {pred_time_knn_orig:.6f} s")
print(f"  PCA      – Accuracy: {acc_knn_pca:.4f}, Prediction time: {pred_time_knn_pca:.6f} s")
print(f"  KBest    – Accuracy: {acc_knn_kb:.4f}, Prediction time: {pred_time_knn_kb:.6f} s")
print()

# --- Plot 7: RandomForest Accuracy ---
plt.figure(figsize=(5, 3))
plt.bar(spaces_3, rf_accs)
plt.ylabel("Accuracy")
plt.ylim(0, 1.0)
plt.title("RandomForest – Accuracy by feature space")
plt.tight_layout()
plt.show()

print("RandomForest summary (printed values):")
print(f"  Original – Accuracy: {acc_rf_orig:.4f}")
print(f"  PCA      – Accuracy: {acc_rf_pca:.4f}")
print(f"  KBest    – Accuracy: {acc_rf_kb:.4f}")
print()

print("Step 6 completed: metrics aggregated and visual comparisons plotted.")
```
